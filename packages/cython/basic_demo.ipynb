```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from feedback.simulation import Points
from feedback.basic import BasicSimulation
```

```python
simulation = BasicSimulation(-3, 3, -3, 3)
```

```python
def test(mp):
    p = Points(x_min=-5, x_max=5, y_min=-5, y_max=5, width=mp * 1000, height=1000)
    simulation.run(p)

for mp in [1, 10, 100]:
    t = %timeit -o -q -r 1 test(mp)
    avg = sum(t.timings) / len(t.timings)
    mp_per_s = mp / avg
    print(f"megapixels per second at {mp} MP: {mp_per_s}")

```

```python
def demo():
    p = Points(x_min=-5, x_max=5, y_min=-5, y_max=5, width=100, height=100)
    simulation.run(p)
    depth = p.data['depth']
    
    plt.imshow(depth, norm=matplotlib.colors.Normalize())
    plt.title("Scene depth")
    plt.colorbar()
    plt.xlabel('x')
    plt.ylabel('y')
demo()
```
